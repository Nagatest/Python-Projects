```python
def even_or_odd(num):
    if num % 2 == 0:
        return "The Entered number {} is Even".format(num)
    else:
        return "The Entered number {} is odd".format(num)
```

```python
even_or_odd(8)
```

```python
lst = [1,2,3,4,5,6,7,8,9,0,11,12,13,14,15,16,17]
list(map(even_or_odd,lst))
```

```python
#Filter Function - applying the filter for fucntion
def even(num):
    if num % 2 == 0:
        return True
```

```python
lst = [1,2,3,4,5,6,7,8,9,0]
list(filter(even,lst))
```

```python
#using the lambda functions
list(filter(lambda num: num%2==0, lst))
```

```python
#list Compreshions
#List Comprehensions provide a concise way to create lists.
lst =[]
def list_square(lst):
    for i in lst:
        lst.append(i*i)
    return lst
```
