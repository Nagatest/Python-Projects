```python
################# Converting appropriate required Datatypes ####################
Absenteesim_Data['Reason for absence'] = Absenteesim_Data['Reason for absence'].replace(0,20)
Absenteesim_Data['Month of absence'] = Absenteesim_Data['Month of absence'].replace(0,np.nan)

#Relacing the null value with median
Absenteesim_Data['Month of absence']=Absenteesim_Data['Month of absence'].fillna(Absenteesim_Data['Month of absence'].median())

#Converting colmnns into categorical factors as they contain unqiue values
category_column_names = ['ID','Reason for absence','Month of absence','Day of the week','Seasons','Disciplinary failure', 'Education', 
                         'Son', 'Social drinker','Social smoker', 'Pet']
for i in category_column_names :
    Absenteesim_Data[i] = Absenteesim_Data[i].astype('category')
    
continous_column_names = []


#Converting rest variables into float64 for standardization
for i in Absenteesim_Data.columns:
    if i not in category_column_names:
        Absenteesim_Data[i] = Absenteesim_Data[i].astype('float')
        continous_column_names.append(i)

print("Categorical Columns : ",category_column_names)
print()
print("Continous Columns : ",continous_column_names)
```

```python
#As per the box plot below are the variable having the outlier
#'Transportation expense','Service time','Age','Work load Average/day ','Hit target','Height','Absenteeism time in hours'
#Capping the outlier
for i in ['Transportation expense','Service time','Age','Work load Average/day ','Hit target','Height','Absenteeism time in hours']:
    q75,q25 = np.percentile(Absenteesim_Data.loc[:,i],[75,25])
    iqr = q75-q25
    min = q25-1.5*iqr
    max = q75+1.5*iqr
    Absenteesim_Data.loc[Absenteesim_Data[i] < min,i] = min
    Absenteesim_Data.loc[Absenteesim_Data[i] > max,i] = max
```

```python
def Relationship_two_Variable(a,b):
    Relation = Absenteesim_Data.groupby(a)[b].sum()
    Relation = Relation.reset_index()
    Relation[b] = (Relation[b]*100)/sum(Absenteesim_Data[b])
    Relation = Relation.sort_values(by=['Absenteeism time in hours'])
    print(Relationship_two_Variable)
    return Relation.plot.barh(x=a,y=b)
```

```python
from scipy import stats
#loop for ANOVA test Since the target variable is continuous
for i in category_column_names:
    f, p = stats.f_oneway(Absenteesim_Data[i], Absenteesim_Data["Absenteeism time in hours"])
    print("P value for variable "+str(i)+" is "+str(p))
```

```python
# Feature Elimination Since it contains redundant Information and No Information ['Weight']
drop_var = ['Weight']
Absenteesim_Data = Absenteesim_Data.drop(drop_var, axis = 1)
```

```python
# Updating the Continuous and Categorical Variables
continous_column_names.remove('Weight')
```

```python
#Normality check
for i in continous_column_names:
    if i == 'Absenteeism time in hours':
        continue
    sns.distplot(Absenteesim_Data[i],bins = 'auto')
    plt.title("Checking Distribution for Variable "+str(i))
    plt.ylabel("Density")
    plt.show()
```

```python
#Normalization of continuous variables
for i in continous_column_names:
    if i == 'Absenteeism time in hours':
        continue
    Absenteesim_Data[i] = (Absenteesim_Data[i] - Absenteesim_Data[i].min())/(Absenteesim_Data[i].max()-Absenteesim_Data[i].min())
```

```python
# Encoding Categorical Data
Absentees_Data = pd.get_dummies(data = Absenteesim_Data, columns = category_column_names)
```
