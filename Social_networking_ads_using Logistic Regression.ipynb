```python
#Random forest
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import os
os.chdir("C:\\Data Science\\social-network-ads\\")
```

```python
social = pd.read_csv("Social_Network_Ads.csv")
social.head()
```

```python
X = social.iloc[:,[2,3]].values
y = social.iloc[:,4].values
```

```python
from  sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.7, test_size =0.3, random_state = 100)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
#Create the model for lOgistic Regression
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression(random_state = 0)
```

```python
lm.fit(X_train,y_train)
```

```python
#Predict 
y_pred = lm.predict(X_test)
```

```python
y_pred
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
mt = confusion_matrix(y_test,y_pred)
mt
```

```python
#Finding the accuracy_score
from sklearn.metrics import accuracy_score
accuracyscore = accuracy_score(y_test,y_pred)
accuracyscore
```

```python
#Classofcation report of the predicted data
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred)
print(report)
```
